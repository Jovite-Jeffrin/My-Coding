```python
# selecting appropriate columns

df = df[['Company Name', 'Model Name', 'Mobile Weight', 'RAM', 'Front Camera',
       'Back Camera', 'Processor', 'Battery Capacity', 'Screen Size',
       # 'Launched Price (Pakistan)',
        'Launched Price (India)',
       # 'Launched Price (China)', 
        'Launched Price (USA)',
       'Launched Price (Dubai)', 'Launched Year']]
```

```python
df.head()
```

```python
df['Company Name'].unique()
```

```python
# convert the year column to numeric

pd.to_numeric(df['Launched Year'])
```

```python
df.dtypes
```

```python
df.rename(columns = {
    'Launched Price (India)' : 'Launched Price (India (INR))',
    'Launched Price (USA)' : 'Launched Price (USA (USD))',
    'Launched Price (Dubai)' : 'Launched Price (Dubai (AED))'
}, inplace = True)
```

```python
df.columns
```

```python
df['Launched Price (India (INR))'] = df['Launched Price (India (INR))']\
                                    .replace(r'[^\d]','',regex=True)\
                                    .astype(int)
```

```python
df.head()
```

```python
df['Launched Price (USA (USD))'] = df['Launched Price (USA (USD))']\
                                    .replace(r'[^\d]','',regex=True)\
                                    .astype(int)
```

```python
df['Launched Price (Dubai (AED))'] = df['Launched Price (Dubai (AED))']\
                                    .replace(r'[^\d]','',regex=True)\
                                    .astype(int)
```
